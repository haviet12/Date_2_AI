```python
import pandas as pd
url= 'http://winterolympicsmedals.com/medals.csv'
data=pd.read_csv(url)
print(data)
```

```python
data_del=data.drop(['City','Sport','Event gender','Event'], axis =1)
data_del.groupby('Medal').count()
## tổng hợp số huy chương vàng, bạc, đồng từ năm 1924-> 2006
```

```python
## danh sách giải thưởng của CANADA qua các năm
frame = data.set_index(data['NOC'])
frame.loc['CAN']
```

```python
frame = data.set_index(data['NOC'])
filter=frame.loc['CAN']
filter.groupby('Medal').count()
## liệt kê tổng số huy chương mà CANADA đạt được 
```

```python
## danh sách liệt kê số huy chương đạt được ở từng hạng mục thi đấu olympic 
filter.groupby(['Sport','Event','Medal']).count()
```
